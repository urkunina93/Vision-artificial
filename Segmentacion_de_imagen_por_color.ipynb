# Segmentacion de imagen por color

three_d_array.png

```python
import numpy as np
import cv2

from google.colab.patches import cv2_imshow  # Importar cv2_imshow

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/25.JPG')  # Cargamos la imagen
height, width = img.shape[:2]  # Tomamos las dimensiones de la imagen cargada
segImg = np.zeros(img.shape, np.uint8)  # Creamos una imagen vacia con las mismas dimensines que la imagen cargada

for i in range(0, height):  # Recorremos la imagen elemento por elemento
    for j in range(0, width):
        if img[i, j, 0] <= 40 and img[i, j, 1] <= 40 and img[i, j, 2] >= 200:  # Limitamos el rango de valores que deseamos obervar
            # Definimos los nuevos valores que tomara cada canal
            segImg[i, j, 0] = 0  # Canal azul
            segImg[i, j, 1] = 0  # Canal verde
            segImg[i, j, 2] = 255  # Canal rojo



cv2_imshow(img)  # Mostrar imagen original
cv2_imshow(segImg)  # Mostrar imagen segmentada


# Fuente: Documentacion OpenCV
```

# segmentacion por rangos de color

```python
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/25.JPG')  # Cargamos la imagen
height, width = img.shape[:2]  # Tomamos las dimensiones de la imagen cargada

lower_red = np.array([0, 0, 200])  # Creamos un vector con los valores minimos del limite
upper_red = np.array([40, 40, 255])  # Creamos un vector con los valores maximos del limite

mask = cv2.inRange(img, lower_red, upper_red)  # Creamos una mascara con los rangos a segmentar

segImg1 = cv2.bitwise_and(img, img, mask=mask)  # Realimos una operacion bit a bit para aplicar la mascara a la imagen final (conjuncioon logica)
segImg2 = cv2.inRange(img, lower_red, upper_red)  # Segemetacion usando el comando inrange

cv2_imshow(img)  # Mostramos a imagen original
print("imagen segementada")
cv2_imshow(segImg1)  # Mostramos la imagen segementada
print("mascara")
cv2_imshow(segImg2)  # Mostramos la mascara de la imagen

# Fuente: Documentacion OpenCV
```
